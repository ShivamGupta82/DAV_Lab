## **Using Python**

### **1. Loading the dataset**

```r
# Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import altair as alt
import pandas as pd

# Load a built-in dataset from seaborn
tips = sns.load_dataset('tips')
tips.head()
```

### **2. Visualizations**

```r
# Creating a simple histogram using Matplotlib
plt.hist(tips['total_bill'], bins=20, color='skyblue')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill')
plt.show()
```

```r
#Creating an interactive scatter plot using Plotly Express
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', hover_name='time', log_x=True)
fig.update_layout(title='Tip vs Total Bill (Interactive)', xaxis_title='Total Bill', yaxis_title='Tip')
fig.show()
```

```r
#Creating a simple interactive chart using Altair
alt.Chart(tips).mark_circle().encode(
    x='total_bill',
    y='tip',
    color='sex',
    tooltip=['total_bill', 'tip', 'sex']
).interactive().properties(
    title='Tip vs Total Bill (Interactive)'
)
```

```r
# Creating a boxplot to compare the distribution of total bill for different days using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Distribution of Total Bill by Day')
plt.show()
```

```r
# Creating a violin plot to visualize the distribution of tip by time and day using Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='tip', hue='time', data=tips, split=True)
plt.xlabel('Day')
plt.ylabel('Tip')
plt.title('Distribution of Tip by Time and Day')
plt.legend(title='Time')
plt.show()
```

```r
# Creating a bar plot to compare the average total bill by size of the group using Plotly Express
fig = px.bar(tips, x='size', y='total_bill', color='size', title='Average Total Bill by Group Size')
fig.update_layout(xaxis_title='Size of Group', yaxis_title='Average Total Bill')
fig.show()
```

```r
# Creating a scatter plot matrix to visualize pairwise relationships in the dataset using Seaborn
sns.pairplot(tips, hue='sex')
plt.suptitle('Pairwise Relationships in the Tips Dataset')
plt.show()
```

```r
# Creating a stacked bar chart to visualize the count of male and female customers by day using Altair
alt.Chart(tips).mark_bar().encode(
    x='day',
    y='count()',
    color='sex'
).properties(
    title='Count of Male and Female Customers by Day'
)
```

## **Using R**

```r
library(ggplot2)
library(leaflet)
library(dplyr)

# Load a built-in dataset
data(mtcars)
```

```r
# Creating a boxplot to compare the distribution of miles per gallon (mpg) for different number of cylinders
ggplot(data = mtcars, aes(x = factor(cyl), y = mpg)) +
  geom_boxplot(fill = "skyblue") +
  labs(x = "Number of Cylinders", y = "Miles Per Gallon", title = "Distribution of MPG by Number of Cylinders")
```

```r
# Creating a scatter plot to visualize the relationship between miles per gallon (mpg) and horsepower (hp)
ggplot(data = mtcars, aes(x = hp, y = mpg)) +
  geom_point(color = "blue", alpha = 0.7) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(x = "Horsepower", y = "Miles Per Gallon", title = "Scatter plot of MPG vs HP")
```

```r
# Creating a histogram to visualize the distribution of horsepower (hp)
ggplot(data = mtcars, aes(x = hp)) +
  geom_histogram(fill = "skyblue", bins = 10) +
  labs(x = "Horsepower", y = "Frequency", title = "Distribution of Horsepower")
```

```r
# Creating a scatter plot to visualize the relationship between miles per gallon (mpg) and weight (wt)
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
  geom_point(color = "blue", alpha = 0.7) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(x = "Weight", y = "Miles Per Gallon", title = "Scatter plot of MPG vs Weight")
```

```r
# Creating a bar plot to compare the average mpg by number of gears
mtcars %>%
  group_by(gear) %>%
  summarise(avg_mpg = mean(mpg)) %>%
  ggplot(aes(x = factor(gear), y = avg_mpg)) +
  geom_bar(stat = "identity", fill = "skyblue") +
  labs(x = "Number of Gears", y = "Average MPG", title = "Average MPG by Number of Gears")
```

```r
# Creating a bar plot to compare the average mpg by number of cylinders
mtcars %>%
  group_by(cyl) %>%
  summarise(avg_mpg = mean(mpg)) %>%
  ggplot(aes(x = factor(cyl), y = avg_mpg)) +
  geom_bar(stat = "identity", fill = "skyblue") +
  labs(x = "Number of Cylinders", y = "Average MPG", title = "Average MPG by Number of Cylinders")
```

```r
# Creating a boxplot to compare the distribution of miles per gallon (mpg) by transmission type (am)
ggplot(data = mtcars, aes(x = factor(am), y = mpg, fill = factor(am))) +
  geom_boxplot() +
  labs(x = "Transmission Type", y = "Miles Per Gallon", title = "Distribution of MPG by Transmission Type")
```

### **Conclusion:**
Relationship between MPG and Weight: We observed a negative relationship between miles per gallon (MPG) and weight, indicating that heavier cars tend to have lower fuel efficiency.

Impact of Number of Cylinders on MPG: The bar plot comparing the average MPG by the number of cylinders revealed that cars with fewer cylinders tend to have higher average MPG.

Distribution of Horsepower: The histogram showed that the distribution of horsepower is skewed to the right, indicating that most cars have moderate to high horsepower.

Effect of Transmission Type on MPG: The boxplot comparing the distribution of MPG by transmission type showed that cars with automatic transmission (am = 0) generally have lower MPG compared to those with manual transmission (am = 1).

Distribution of MPG by Number of Gears: The density plot illustrated the distribution of MPG by the number of gears, providing insights into the variability in fuel efficiency across different gear configurations.
