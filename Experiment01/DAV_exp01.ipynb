### **1. Numpy**

```python
import numpy as np
import sys

# Converting python list to array

myList  = [1,2,3,4,5]
myArray = np.array(myList)
print(f"Python Array : {myList} type:{type(myList)} size:{sys.getsizeof(myList)} bytes\nNumpy Array:{myArray} type:{type(myArray)} size:{myArray.size} bytes")
```

```python
# Creating 1D,2D arrays

array1D = np.array([2,3,4,5])
print(f"1D Array : {array1D}, Shape : {array1D.shape}")

array2D = np.array([[1,2,3],[4,5,6]])
print(f"2D Array : \n{array2D}, Shape : {array2D.shape}")
```

```python
#Creating 3D array and  one method , zero , arrange

shape = (2,3,4)
one = np.ones(shape)
print(one)

zero = np.zeros(shape)
print(zero)
```

```python
#using reshape

print(f"Before reshape :\n{one}")
#reshape to 2d array
print(f"\nAfter reshape :\n{one.reshape(6,4)}")
#reshape to 3d array
print(f"\nAfter reshape :\n {one.reshape(-1,1)}")
```

### **3. Scikit**

```python
from sklearn.linear_model import LinearRegression

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
print(X)
y = np.dot(X, np.array([1, 2])) +3
print(y)
reg = LinearRegression().fit(X, y)
reg.score(X, y)
reg.coef_
reg.predict(np.array([[3, 5]]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

diabetes_X = diabetes_X[:, np.newaxis, 2]

diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

regr = linear_model.LinearRegression()

regr.fit(diabetes_X_train, diabetes_y_train)

diabetes_y_pred = regr.predict(diabetes_X_test)

print("Score: \n", regr.score(diabetes_X_test,diabetes_y_test))
print("Coefficients: \n", regr.coef_)
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

## **R Data Libraries**

### **1. Dplyr**

```python
library(dplyr)
install.packages("gapminder")
library(gapminder)
```

```python
head(gapminder)
```

```python
# Select
gapminder %>%
  select(country, year, pop)
```

```python
# Select with minus
gapminder %>%
  select(-continent)
```

```python
# filter
gapminder %>%
  select(country, year, lifeExp) %>%
  filter(year == 2007)
```

```python
# filter with multiple parameters
gapminder %>%
  select(country, year, lifeExp) %>%
  filter(year == 2007, country == "Poland")
```

```python
# check for multiple values using %in%
gapminder %>%
  select(country, year, lifeExp) %>%
  filter(year == 2007, country %in% c("Poland", "Croatia"))
```

```python
# arrange (like sorting within a column)
gapminder %>%
  select(country, year, lifeExp) %>%
  filter(year == 2007) %>%
  arrange(lifeExp)
```

```python
# descending order
gapminder %>%
  select(country, year, lifeExp) %>%
  filter(year == 2007) %>%
  arrange(desc(lifeExp))
```

```python
gapminder %>%
  select(country, year, lifeExp) %>%
  filter(year == 2007) %>%
  arrange(desc(lifeExp)) %>%
  top_n(5)
```

```python
# create new attribute using mutate
gapminder %>%
  select(country, year, pop, gdpPercap) %>%
  filter(year == 2007) %>%
  mutate(gdp = pop * gdpPercap) %>%
  arrange(desc(gdp)) %>%
  top_n(5)
```

```python
# transmutate
gapminder %>%
  select(country, year, pop, gdpPercap) %>%
  filter(year == 2007) %>%
  transmute(gdp = pop * gdpPercap) %>%
  arrange(desc(gdp)) %>%
  top_n(5)
```

```python
# average of a col
gapminder %>%
  summarize(avgLifeExp = mean(lifeExp))
```

```python
gapminder %>%
  filter(year == 2007, continent == "Europe") %>%
  summarize(avgLifeExp = mean(lifeExp))
```

```python
# group_by
gapminder %>%
  filter(year == 2007) %>%
  group_by(continent) %>%
  summarize(avgLifeExp = mean(lifeExp))
```

```python
rename(gapminder,continents = continent) %>% slice_sample(n = 10)
```

```python
ggplot(data_iris) +
  geom_point(aes(x = Sepal.Length , y = Sepal.Width),
             color = 'blue', size = 4, pch = 19) +

  ggtitle("Basic Scatter Plot", subtitle = "iris") +
  xlab("SEPAL LENGTH") +
  ylab("SEPAL WIDTH")
```

```python
ggplot(data_iris) +
  geom_point(aes(x = Sepal.Length , y = Sepal.Width,
                 color = Species, shape = Species), size = 4) +

  ggtitle("Scatter Plot with feature differentiation - 1") +
  xlab("SEPAL LENGTH") +
  ylab("SEPAL WIDTH") +
  theme_bw() +
  theme(axis.text.x = element_text(face = 'bold.italic',
                                   color = 'darkgreen',
                                   size = 10, angle = 0),
        axis.text.y = element_text(face = 'bold',
                                   color = 'blue',
                                   size = 10, angle = 45))
```

```python
ggplot(data_iris) +
  geom_point(aes(x = Sepal.Length , y = Sepal.Width,
                 size = Petal.Width, alpha = Petal.Length)) +

  ggtitle("Scatter Plot with feature differentiation - 2") +
  xlab("SEPAL LENGTH") +
  ylab("SEPAL WIDTH") +
  theme_bw() +
  theme(axis.text.x = element_text(face = 'bold.italic',
                                   color = 'darkgreen',
                                   size = 10, angle = 0),
        axis.text.y = element_text(face = 'bold',
                                   color = 'blue',
                                   size = 10, angle = 45))
```

```python
ggplot(data_iris) +
  geom_line(aes(x = Sepal.Length, y = Sepal.Width),
            lwd = 1.25, color = 'darkgreen') +

  ggtitle("Basic Line Plot") +
  xlab("Sepal Length") +
  ylab("Sepal Width") +
  theme_bw() +
  theme(axis.text.x = element_text(face = 'bold', size = 10),
        axis.text.y = element_text(face = 'bold', size = 10))
```

```python
ggplot(data_iris) +
  geom_bar(aes(x = Sepal.Length, y = Sepal.Width),
           stat = 'identity', fill = 'darkgreen') +

  ggtitle("Basic Bar Plot") +
  xlab("Sepal Length") +
  ylab("Sepal Width") +
  theme_bw() +
  theme(axis.text.x = element_text(face = 'bold', size = 10),
        axis.text.y = element_text(face = 'bold', size = 10))
```

```python
# box plot
ggplot(iris, aes(x = Species, y = Sepal.Length, fill = Species)) +
  geom_boxplot() +
  ggtitle("Boxplot of Sepal Length by Species") +
  xlab("Species") +
  ylab("Sepal Length") +
  theme_bw() +
  theme(axis.text.x = element_text(face = 'bold', size = 10),
        axis.text.y = element_text(face = 'bold', size = 10))
```

```python
# polar plot
ggplot(iris, aes(x = Sepal.Length, y = 1, color = Species)) +
  geom_point(stat = "identity", position = "identity", size = 3) +
  coord_polar(theta = "x") +
  ggtitle("Polar Plot of Sepal Length by Species") +
  theme_minimal()
```

```python
# violin plot
ggplot(iris, aes(x = Species, y = Sepal.Length, fill = Species)) +
  geom_violin() +
  ggtitle("Violin Plot of Sepal Length by Species") +
  xlab("Species") +
  ylab("Sepal Length") +
  theme_minimal() +
  theme(axis.text.x = element_text(face = 'bold', size = 10),
        axis.text.y = element_text(face = 'bold', size = 10))
```
